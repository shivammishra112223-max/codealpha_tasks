```python
df['ProfileName']=df['ProfileName'].fillna(df['ProfileName'].mode()[0])
df['Summary']=df['Summary'].fillna(df['Summary'].mode()[0])


```

```python
print(df.isnull().sum())
```

#Outlier chek

```python
cols = ['HelpfulnessNumerator', 'HelpfulnessDenominator']

for col in cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1
    lower = q1 -1.5*IQR 
    upper = q3 + 1.5*IQR
    
    outlier = df[(df[col] < lower) | (df[col] > upper)]
    print(col, "Outliers:", len(outlier))
```

```python
df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
print("duplicate value",df.duplicated().sum())
```

#Matrics Banana

```python
top_users = df['UserId'].value_counts().head(5000).index

top_products = df['ProductId'].value_counts().head(5000).index
```

```python
df_filtered = df[
    df['UserId'].isin(top_users) &
    df['ProductId'].isin(top_products)
]
```

```python
matrix = df_filtered.pivot_table(
    index='UserId',
    columns='ProductId',
    values='Score'
).fillna(0)
```
