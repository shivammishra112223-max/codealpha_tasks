```python
import  numpy as py
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("C:\\Users\\shiva\\OneDrive\\Pictures\\Documents\\Unemployment_Rate_upto_11_2020.csv")
print(Data.head(5))
```

```python
print(Data.isnull().sum())
```

```python
print(Data.info())
```

```python
Data.columns = Data.columns.str.strip()
Data['Date']=pd.to_datetime(Data['Date'])
print(Data.info())                          
```

```python
import matplotlib.pyplot as plt

plt.plot(Data['Date'], Data['Estimated Unemployment Rate (%)'])
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")
plt.title("Unemployment Trend")
plt.show()
```

```python
Before =Data[Data['Date'] < '2020-01-01']
during = Data[(Data['Date'] >= '2020-01-01') & (Data['Date'] <= '2020-12-31')]
after = Data[Data['Date'] > '2020-12-31']

plt.plot(Before['Date'],Before['Estimated Unemployment Rate (%)'],label="before covid")
plt.plot(during['Date'], during['Estimated Unemployment Rate (%)'], label="During Covid")
plt.plot(after['Date'], after['Estimated Unemployment Rate (%)'], label="After Covid")


plt.legend()
plt.title("Covid Impact on Unemployment")
plt.show()
```

```python
Data['Month'] = Data['Date'].dt.month

monthly = Data.groupby('Month')['Estimated Unemployment Rate (%)'].mean()


monthly.plot()
plt.title("Monthly Unemployment Trend")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.show()
```
